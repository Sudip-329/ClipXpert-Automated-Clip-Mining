```python
import yt_dlp as youtube_dl
from pydub import AudioSegment

video_url = 'https://www.youtube.com/watch?v=05qZCsS6qEw'

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'outtmpl': 'downloaded_audio.%(ext)s'
}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])

audio = AudioSegment.from_file('downloaded_audio.mp3', format='mp3')

segment = audio[:10000] 
segment.export('audio_segment_10sec.mp3', format='mp3')

audio.export('downloaded_audio_wav_format.wav', format='wav')
```

# Downloading Subtitle from YT (Auto Generated Title)

```python

import yt_dlp as youtube_dl
video_url = 'https://www.youtube.com/watch?v=05qZCsS6qEw'

ydl_opts = {
    'writesubtitles': True,
    'subtitleslangs': ['en'], 
    'writeautomaticsub': True,  
    'skip_download': True,  
    'outtmpl': 'downloaded_subtitles.%(ext)s' 
}

with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])
```

```python
# Display the contents of the subtitle file
with open('downloaded_subtitles.en.vtt', 'r') as file:
    subtitles = file.read()

print(subtitles)
```

# Removing Time SLice, Extra Characters and duplicate lines

```python
import re
subtitle_file = 'downloaded_subtitles.en.vtt'
with open(subtitle_file, 'r') as file:
    subtitles = file.read()

subtitles = re.sub(r'WEBVTT.*\n', '', subtitles)

subtitles = re.sub(r'\d{2}:\d{2}:\d{2}\.\d{3} --> \d{2}:\d{2}:\d{2}\.\d{3}.*\n', '', subtitles)

subtitles = re.sub(r'<.*?>', '', subtitles)

lines = subtitles.split('\n')
unique_lines = []
seen = set()

for line in lines:
    stripped_line = line.strip()
    if stripped_line and stripped_line not in seen:
        unique_lines.append(stripped_line)
        seen.add(stripped_line)

cleaned_subtitles = '\n'.join(unique_lines)

plain_text_file = 'subtitles.txt'
with open(plain_text_file, 'w') as file:
    file.write(cleaned_subtitles)

with open(plain_text_file, 'r') as file:
    plain_text = file.read()

print(plain_text)
```

```python
import whisper
from langdetect import detect, LangDetectException

model = whisper.load_model("base")
audio_path = '/content/downloaded_audio.mp3'
result = model.transcribe(audio_path)
transcription = result["text"]

def filter_english_text(text):
    sentences = text.split('.')
    english_sentences = []
    for sentence in sentences:
        try:
            if detect(sentence.strip()) == 'en':
                english_sentences.append(sentence.strip())
        except LangDetectException:
            continue
    return '. '.join(english_sentences)

english_transcription = filter_english_text(transcription)

output_file = '/content/openai_whisper_subtitle.txt'
with open(output_file, 'w') as file:
    file.write(english_transcription)

print(english_transcription)
```
